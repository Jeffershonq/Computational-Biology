```python
# Import Libraries

from Bio.Seq import Seq
from Bio import SeqIO

seq1 = SeqIO.read('./KY352407.fasta', 'fasta').seq
seq2 = SeqIO.read('./OQ297732.fasta', 'fasta').seq

print(seq1)
print(seq2)
```

```python
# Total Length

seq1len = len(seq1)
seq2len = len(seq2)

print(seq1len)
print(seq2len)
```

```python
# Codon

codon = 'CGC'

print(seq1.find(codon))
print(seq2.find(codon))
```

```python
# Neulotides

slicing1 = seq1[:6]
slicing2 = seq2[-9:]

# combine
seq3 = slicing1 + slicing2
print(seq3)
```

Sequence Analysis & Plotting

```python
ambigous = ['N', 'Y', 'R', 'W', 'M', 'S', 'K']

for i in ambigous:
    seq1 = seq1.replace(i, '')
    seq2 = seq2.replace(i, '')
```

```python
from Bio.SeqUtils import MeltingTemp as mt

wall1 = mt.Tm_Wallace(seq1)
wall2 = mt.Tm_Wallace(seq2)

print(wall1)
print(wall2)
```

```python
from Bio.SeqUtils import GC

gc1 = GC(seq1)
print('Persentage GC: ',gc1,'%')

gc2 = GC(seq2)
print('Persentage GC: ',gc2,'%')

gc3 = GC(seq3)
print('Persentage GC: ',gc3,'%')
```

```python
at1 = 100 - gc1
print('Persentage AT: ',at1,'%')

at2 = 100 - gc2
print('Persentage AT: ',at2,'%')

at3 = 100 - gc3
print('Persentage AT: ',at3,'%')
```

```python
from Bio.SeqUtils import molecular_weight as mw

mw1 = mw(seq1)
print(mw1)

mw2 = mw(seq2)
print(mw2)

mw3 = mw(seq3)
print(mw3)
```

```python
from collections import Counter

counter1 = Counter(seq1)
key1 = counter1.keys()
value1 = counter1.values()

print(counter1)
print(key1)
print(value1)

counter2 = Counter(seq2)
key2 = counter2.keys()
value2 = counter2.values()

print(counter2)
print(key2)
print(value2)

counter3 = Counter(seq3)
key3 = counter3.keys()
value3 = counter3.values()

print(counter3)
print(key3)
print(value3)
```

```python
from matplotlib import pyplot as plt

plt.bar(key1,value1)
plt.show()

plt.bar(key2,value2)
plt.show()

plt.bar(key3,value3)
plt.show()
```

```python
plt.title("Pie Chart Mode")
plt.pie(counter1.values(), labels = counter1.keys())
plt.show()

plt.title("Another Pie")
plt.pie(counter2.values(), labels = counter2.keys())
plt.show()
```

DNA & mRNA Protein Synthesis

```python
mRNA1 = seq1.transcribe()
print(mRNA1)

mRNA2 = seq2.transcribe()
print(mRNA2)

mRNA3 = seq3.transcribe()
print(mRNA3)
```

```python
# Translate

amino1 = mRNA1.translate()
amino2 = mRNA2.translate()
amino3 = mRNA3.translate()

print(amino1)
print(amino2)
print(amino3)
```
