```python
# Import Libraries

from Bio.Seq import Seq

seq1 = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
seq2 = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
```

```python
from Bio.SeqUtils import molecular_weight as mw
from Bio.SeqUtils import MeltingTemp as mt
from Bio.SeqUtils import GC

mw1 = mw(seq1)
print(mw1)

mw2 = mw(seq2)
print(mw2)
```

```python
melting1 = mt.Tm_Wallace(seq1)
print(melting1)

melting2 = mt.Tm_Wallace(seq2)
print(melting2)

from matplotlib import pyplot as plt
plt.title("Melting Temp")
plt.bar(['MT A', 'MT B'],[melting1,melting2])
plt.show()
```

```python
# Calculate GC Content

gcSeq1 = GC(seq1)
print(gcSeq1)

gcSeq2 = GC(seq2)
print(gcSeq2)
```

```python
# AT

atSeq1 = 100 - gcSeq1
print(atSeq1)

atSeq2 = 100 - gcSeq2
print(atSeq2)

plt.title("AT")
plt.bar(['AT A', 'AT B'],[atSeq1, atSeq2])
plt.show()
```
