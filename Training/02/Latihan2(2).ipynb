```python
from Bio.Seq import Seq

seq1 = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
seq2 = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
print(seq1)
print(seq2)
```

```python
from Bio.SeqUtils import molecular_weight as mw

mw1 = mw(seq1)
print(mw1)

mw2 = mw(seq2)
print(mw2)
```

```python
from Bio.SeqUtils import MeltingTemp as mt

mt1 = mt.Tm_Wallace(seq1)
print(mt1)

mt2 = mt.Tm_Wallace(seq2)
print(mt2)
```

```python
from Bio.SeqUtils import GC

gc1 = GC(seq1)
print(gc1)

gc2 = GC(seq2)
print(gc2)
```

```python
at1 = 100 - gc1
at2 = 100 - gc2

print(at1)
print(at2)
```

```python
from matplotlib import pyplot as plt

plt.bar(['GC A', 'AT A'], [gc1, at1])
plt.show()
plt.bar(['GC B', 'AT B'], [gc2, at2])
plt.show()
```
