```python
import Bio
from Bio.Seq import Seq
from Bio.SeqUtils import nt_search
import matplotlib.pyplot as plt
from collections import Counter
```

```python

seqA = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
seqB = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
```

```python
#LATIHAN SOAL CASE 1
#•	Find the length of both sequences.

print(f'Length of A = {len(seqA)}')
print(f'Length of B = {len(seqB)}')
print(f'Length of A + B = {len(seqA + seqB)}')

#•	Determine the number of times the codon (CGC) appears in both sequences.

print(f"\nCGC Appearance amount of Seq A is = {seqA.count('CGC')}")
print(f"CGC Appearance amount of Seq B is = {seqB.count('CGC')}")
print(f"Amount of appearance using nt_search = {nt_search(str(seqA), 'CGC')}") 

#•	Locate the first occurrence index position of the sub-sequence (CAGTC) in both sequences.

print(f"\nFirst Sequence of CAGTC of Seq A is = {seqA.index('CAGTC')}")
print(f"First Sequence of CAGTC of seq B is = {seqB.find('CAGTC')}")

#•	Combine the first 15 nucleotides from sequence A and the last 10 nucleotides from sequence B to create a new DNA sequence called sequence C.

print(f'\n Combined nucleotides sequence is = {seqA[:15] + seqB[-10:]}')

#•	Reverse the sequence of sequence C.

seqC = (seqA[:15] + seqB[-10:])
print(f'The sequence of Seq C is = {seqC}')


#•	Plot the nucleotide base (A, C, G, T) frequency of each of the sequence A, sequence B, and sequence C using the Matplotlib library.
counterA = Counter(seqA)
counterB = Counter(seqB)
counterC = Counter(seqC)

plt.bar(counterA.keys(), counterA.values())
plt.title("Sequence A")
plt.show()

plt.bar(counterB.keys(), counterB.values())
plt.title("Sequence B")
plt.show()

plt.bar(counterC.keys(), counterC.values())
plt.title("Sequence C")
plt.show()
```
