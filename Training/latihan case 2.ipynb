```python
import Bio
from Bio.Seq import Seq
from Bio.SeqUtils import molecular_weight as mw
from Bio.SeqUtils import MeltingTemp as mt
from Bio.SeqUtils import GC
import matplotlib.pyplot as plt
```

```python
seqA = Seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
seqB = Seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
```

```python
#•	Determine the molecular weight of both sequence.

print(f'Molecular weight of Seq A = {mw(seqA)}')
print(f'Molecular weight of Seq B = {mw(seqB)}')

#•	Refer to the Wallace rule to calculate the melting point of sequence A and sequence B and plot them using Matplotlib library.

print(f'\nMelting Temp of Seq A using Wallace is = {mt.Tm_Wallace(seqA)}')
print(f'Melting Temp of Seq B using Wallace is = {mt.Tm_Wallace(seqB)}')

plt.bar(['MT A', 'MT B'], [mt.Tm_Wallace(seqA), mt.Tm_Wallace(seqB)])
plt.show()

#•	Calculate the GC content of sequence A and sequence B.

print(f'\nGC Content Seq A is = {GC(seqA)}')
print(f'GC Content Seq B is = {GC(seqB)}')

print(f'\nAT Content Seq A is = {100 - GC(seqA)}')
print(f'AT Content Seq B is = {100 - GC(seqB)}')

atA = 100 - GC(seqA)
atB = 100 - GC(seqB)

plt.bar(['GC A', 'GC B'], [GC(seqA), GC(seqB)])
plt.show()

plt.bar(['AT A', 'AT B'], [atA, atB])
plt.show()

#•	Calculate the AT content of sequence A and sequence B.
#•	Plot the GC and AT content of sequence A and sequence B using Matplotlib library.
```
