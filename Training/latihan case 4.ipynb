```python
import Bio
from Bio.Seq import Seq
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
import matplotlib.pyplot as plt
```

```python
seqA = Seq('AAGCTTAGCTAGAGCAGTCAGCTAGCTAGCT')
seqB = Seq('TCTGAAGCTAGCTAGTGCATAGCTAGCTAGC')
```

```python
#•	Determine score of best local alignment.

#score_global_alignment = pairwise2.align.globalxx(seqA, seqB, one_alignment_only=True, score_only=True)
score_local_alignment = pairwise2.align.localxx(seqA, seqB,  one_alignment_only=True, score_only=True)
print(score_local_alignment)

#•	Calculate similarity of DNA sequence A and sequence B using hamming distance method.

def hemming_distance(seq1, seq2):
    return len([(x, y) for x, y in zip(seq1, seq2) if x != y])
print(hemming_distance(seqA,seqB))

#•	Calculate similarity of DNA sequence A and sequence B using levenshtein distance method

from Levenshtein import distance
print(distance(seqA, seqB))

#•	Map DNA sequence A and sequence B similarity using dot plot technique.
plt.bar(['Similarity Hemming', 'Similarity Levenshtein'],[hemming_distance(seqA,seqB), distance(seqA,seqB)])
plt.show()
```
