Sequence Manipulation

```python
# First Step (parse)

from Bio import SeqIO

for x in SeqIO.parse("./sequencefasta.fasta", "fasta"):
    print("Implementing Parse")
    print(x.id)
    print(x.name)
    sequence = x.seq
    print(sequence)
    print(" ")

# Second Step (read)

# file = SeqIO.read("./sequence.gb", "genbank")
# sequence1 = file.seq
# More Convenient
sequence1 = SeqIO.read("./sequence.gb", "genbank").seq

# Convert Sequence into String
print("Implementing Read")
str(sequence1)
```

```python
# Taking the first 10 character on Sequence
# Parameter Indicator
# [star:end:jump]

slicing = sequence1[ :10: ]
print(slicing)
```

```python
# Reverse Sequence

reverse = slicing[::-1]
print(reverse)
```

```python
# Join (Combining Sequence with Another Sequence)

seq_join = slicing + reverse
print(seq_join)
```

```python
# Replace (A -> T)

seq_replace = str(seq_join).replace("A", "T")
print(seq_replace)
seq_replace = str(seq_join).replace("T", "A")
print(seq_replace)
```

Sequence Analysis

```python
# Melting Point

from Bio.SeqUtils import MeltingTemp as mt

# Wallace
melting_temp = mt.Tm_Wallace(sequence1)
print("Melting Temp Sequence1 Wallace = ", melting_temp)
# GC
melting_GC = mt.Tm_GC(sequence1)
print("Melting Temp Sequence1 GC = ", melting_GC)
# NN
melting_NN = mt.Tm_NN(sequence1)
print("Melting Temp Sequence1 NN = ",melting_NN)
```

```python
# GC & AT Content

from Bio.SeqUtils import GC

gc = GC(sequence1)
print(gc)

at1 = 100 - GC(sequence1)
# Cara 2
at2 = 100 - gc
print("Cara Pertama: ",at1)
print("Cara Kedua: ", at2)
```

```python
# Molecular Weight

from Bio.SeqUtils import molecular_weight as mw

molecularWeight = mw(sequence1)
print("Molecular Weight of Sequence1: ",molecularWeight)
```

```python
# Frequency Nucleotides

from collections import Counter

counter = Counter(sequence1)
key = counter.keys()
value = counter.values()
print(counter)
print(key)
print(value)
```

```python
# Plotting

from matplotlib import pyplot as plt
plt.bar(key, value)
plt.show()
```

Transcription Converting DNA -> mRNA

```python
# DNA ->  mRNA (Transcribe)

mrna = slicing.transcribe()
print("mRNA", mrna)
```

Translation mRNA -> Amino Acid

```python
asam_amino = mrna.translate()
print("Amino Acid: ",asam_amino)
```

```python
# Non Quiz Topics

complement = slicing.complement()
print("Complement: ", complement)
```

```python
from Bio.Seq import Seq

# Making new Sequence
seq2 = Seq("GCATGCATGGCAT")

#Check the seq2 Length
length = (len(seq2))
print("Length from Sequence 2: ", length)
```

Local Alignment

```python
from Bio import pairwise2

pairwise2.align.localxx(sequence1, seq2)
# Taking the Score only
print("Score Only: ")
pairwise2.align.localxx(sequence1, seq2, one_alignment_only = True, score_only = True)
```

Global Alignment

```python
pairwise2.align.globalxx(sequence1, seq2)
# Taking the Score only
print("Score Only: ")
pairwise2.align.globalxx(sequence1, seq2, one_alignment_only = True, score_only =True)
```

Hamming Distance

```python
hamming_distance = len([(x,y) for x,y in zip(sequence1, seq2) if x != y])
print(hamming_distance)
```

Levensthein

```python
from Levensthein import distance

distance(sequence1,seq2)
```
